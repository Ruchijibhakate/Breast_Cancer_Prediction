# Objective : To develop a predictive model that accurately determines whether a patient has malignant (M) or benign (B) breast cancer based on clinical and histopathological data.

# Domain Analysis:
## Breast Cancer Overview :
Breast cancer is the most common cancer among women worldwide. It arises from the cells of the breast and can spread to other parts of the body. Early detection and treatment are crucial for improving survival rates.
## Key Statistics :
- Incidence: Breast cancer accounts for about 25% of all cancers in women.
- Mortality: It is the second leading cause of cancer death in women, following lung cancer.
- Risk Factors: Age, family history, genetic mutations (e.g., BRCA1/BRCA2), hormonal factors, and lifestyle choices (e.g., alcohol consumption, obesity).
## Types of Breast Cancer :
- Breast cancer is classified into several subtypes based on molecular and histological characteristics:
- Luminal A: Hormone receptor-positive, HER2-negative, low-grade tumors.
- Luminal B: Hormone receptor-positive, HER2-positive or high-grade tumors
- HER2-positive: High levels of HER2 protein, regardless of hormone receptor status.
- Triple-negative: Lack of estrogen receptors, progesterone receptors, and HER2 protein.
- Basal-like: Often overlaps with triple-negative breast cancer and has a higher risk of recurrence.
## Diagnostic Methods :
- Mammography: X-ray imaging of the breast to detect tumors.
- Ultrasound: Uses sound waves to create images of breast tissue.
- MRI: Magnetic resonance imaging for detailed images of breast tissue.
- Biopsy: Removal of tissue samples for microscopic examination.
## Treatment Options :
- Surgery: Lumpectomy, mastectomy, and lymph node removal.
- Radiation Therapy: Uses high-energy rays to kill cancer cells.
- Chemotherapy: Drugs to kill cancer cells or stop them from growing.
- Hormone Therapy: Treats hormone receptor-positive breast cancers.
- Targeted Therapy: Drugs that target specific characteristics of cancer cells, such as HER2.
- Immunotherapy: Boosts the body's immune system to fight cancer.
## Prognostic Factors :
-Tumor Size and Grade: Larger, higher-grade tumors have a worse prognosis.
- Lymph Node Involvement: Presence of cancer in lymph nodes indicates a higher risk of spread.
- Hormone Receptor Status: Hormone receptor-positive cancers generally have a better prognosis.
- HER2 Status: HER2-positive cancers can be more aggressive but may respond well to targeted therapies.
## Research and Advances :
- Genomic Profiling: Identifying genetic mutations that drive cancer growth and response to treatment.
- Personalized Medicine: Tailoring treatment based on individual genetic and molecular profiles.
- AI and Machine Learning: Enhancing diagnostic accuracy and predicting treatment outcomes.
## Challenges :
- Data Privacy: Ensuring patient data is protected and compliant with regulations.
- Access to Care: Addressing disparities in screening and treatment availability.
- Public Awareness: Educating the public about the importance of early detection and regular screenings.
## Conclusion :
Understanding the various aspects of breast cancer, from diagnosis to treatment and prognosis, is essential for developing effective strategies to combat the disease. By leveraging advanced technologies and personalized approaches, we can improve outcomes and provide better care for patients.

## Univariate Analysis Conclusions (Graph-Based)

### 1. diagnosis
- **Type**: Categorical
- **Distribution**: Binary with two classes â€” likely 'M' (Malignant) and 'B' (Benign).
- **Insight**: The dataset is slightly imbalanced with more benign cases.
- **Action**: Consider applying balancing techniques (e.g., SMOTE) if necessary for model fairness.

---

### 2. radius_mean
- **Skewness**: Slightly right-skewed.
- **Distribution**: Majority values between 10â€“20; a few outliers above 25.
- **Insight**: Higher radius tends to indicate malignancy.
- **Action**: Normalize for consistent scaling.

---

### 3. texture_mean
- **Skewness**: Slight positive skew.
- **Distribution**: Concentrated in the 15â€“25 range.
- **Insight**: Texture values moderately vary across classes.
- **Action**: Keep as is; no extreme outliers.

---

### 4. perimeter_mean
- **Distribution**: Similar to radius; right-skewed with some high-end outliers.
- **Insight**: Correlates strongly with radius_mean (Pearson ~1.00).
- **Action**: Might consider feature reduction due to multicollinearity.

---

### 5. area_mean
- **Skewness**: Right-skewed.
- **Insight**: Large area values likely associated with malignant tumors.
- **Action**: Standardize to reduce effect of scale.

---

### 6. smoothness_mean
- **Distribution**: Slight positive skew.
- **Insight**: Majority values are tightly packed, suggesting low variance.
- **Action**: Possibly low predictive power alone.

---

### 7. compactness_mean & concavity_mean
- **Skewness**: Highly right-skewed.
- **Insight**: High values often linked to malignancy.
- **Action**: Log transformation might improve model performance.

---

### 8. concave_points_mean
- **Distribution**: Right-skewed with clear separation in higher values.
- **Insight**: Very predictive feature; higher values for malignant cases.
- **Action**: High priority for model input.

---

### 9. symmetry_mean
- **Skewness**: Mild.
- **Insight**: Values do not separate classes distinctly.
- **Action**: Use in combination with others.

---

### 10. fractal_dimension_mean
- **Distribution**: Uniform and narrow.
- **Insight**: Limited variability; may contribute less to prediction.
- **Action**: Consider during feature selection.

---

### radius_se, texture_se, perimeter_se
- **Insight**: Highly variable with significant skew.
- **Observation**: Many extreme values; may require normalization.
- **Action**: Can benefit from transformation (e.g., Box-Cox).

---

### area_worst, perimeter_worst, radius_worst
- **Distribution**: Right-skewed; clear outliers present.
- **Insight**: Strongly predictive of malignancy.
- **Action**: Normalize and retain.

---

### fractal_dimension_worst
- **Insight**: Among the least variable features.
- **Action**: May not be a strong predictor.

---

## Summary:
- Most **mean** features show right-skewed distributions with outliers.
- Several features (especially `radius`, `perimeter`, `area`, `concave_points`) show strong predictive power based on graphical separation between malignant and benign.

```python
## Creating independent and dependent variable
X = df.drop('diagnosis', axis=1)#independent variable 
y = df.diagnosis#dependent variable 
```

```python
from collections import Counter# importing counter to check count of each label
from imblearn.over_sampling import SMOTE #for balancing the data
sm=SMOTE()#object creation
print(Counter(y))# checking count for each class 
X_sm,y_sm=sm.fit_resample(X,y)#applying sampling on target variable 
print(Counter(y_sm))# checking count after sampling for  each class
```
